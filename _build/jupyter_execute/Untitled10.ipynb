```python
stats.mannwhitneyu(clouds[clouds['status']=='Seeded']['rainfall'],
                   clouds[clouds['status']=='Unseeded']['rainfall'],alternative='greater')
```

```python
stats.ttest_ind(clouds[clouds['status']=='Seeded']['rainfall'],
                clouds[clouds['status']=='Unseeded']['rainfall'],alternative='greater')
```
